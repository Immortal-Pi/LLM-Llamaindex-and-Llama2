```python
from langchain_community.document_loaders import WebBaseLoader

loader=WebBaseLoader("https://www.sandhills.com/careers-and-internships/details/careers/sandhills/1198/business-analyst-intern")
page_data=loader.load().pop().page_content
print(page_data)
```

```python
from langchain_core.prompts import PromptTemplate
```

```python
prompt_extract=PromptTemplate.from_template(
    """
    ### SCRAPED TEXT FROM WEBSITE:
    {page_data}
    ### INSTRUCTION:
    The scraped text is from the career's page of a website.
    Your job is to extract the job postings and return them in JSON format containing 
    following keys: 'role','experience','skills' and 'description'.
    Only returns the valid JSON.
    ### VALID JSON (NO PREAMBLE):
    """
)
chain_extract=prompt_extract | llm
res=chain_extract.invoke(input={'page_data':page_data})
print(res.content)
```

```python
from langchain_core.output_parsers import JsonOutputParser

json_parser=JsonOutputParser()
json_res=json_parser.parse(res.content)
type(json_res)
```

```python
import pandas as pd
df=pd.read_csv('my_portfolio.csv')
df
```

```python
import uuid
import chromadb
client=chromadb.PersistentClient('vectorstore')
collection=client.get_or_create_collection(name='portfolio')
if not collection.count():
    for _,row in df.iterrows():
        collection.add(documents=row['Techstack'],
                      metadatas={'links':row['Links']},
                      ids=[str(uuid.uuid4())]
                      )
```

```python
links=collection.query(query_texts=['Experience in python','Experience in TensorFlow'],
                      n_results=2
                      ).get('metadatas',[])
```

```python
prompt_email=PromptTemplate.from_template(
    """
    ### JOB DESCRIPTION:
    {job_description}

    ###INSTRUCTION:
    You are Amruth, a Business Analyst doing his masters in Business Analytics and AI at University of Texas at Dallas. 
    You have exceptional skills in AI w.r.t deep learning, building LLM models and data science related projects.
        Your job is to write a cold email to the recruiter regarding the job mentioned above describing the capability of Amruth 
        in fulfilling their needs.
        Also add the most relevant ones from the following links to showcase Amruth's portfolio: {link_list}
        Remember you are Amruth, Masters student at University of Texas at Dallas. 
        Do not provide a preamble.
        ### EMAIL (NO PREAMBLE):
    """
)
chain_email=prompt_email | llm
res=chain_email.invoke({'job_description':str(job),'link_list':links})
print(res.content)
```
